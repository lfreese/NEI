```python
def import_and_edit_EPAobs(path):
    '''Import EPA observational data and drop un-needed columns, round Latitude and Longitude, convert to datetime, select only one of the SO2 standards, convert ozone to ppb'''
    EPA_obs_df = pd.concat(map(pd.read_csv, glob.glob(path)))
    EPA_obs_df['date'] = pd.to_datetime(EPA_obs_df['Date Local'])
    EPA_obs_df['date'] = EPA_obs_df['date'].dt.normalize() + timedelta(hours=12)
    EPA_obs_df['Longitude'] = np.round(EPA_obs_df['Longitude'], decimals = 8)
    EPA_obs_df['Latitude'] = np.round(EPA_obs_df['Latitude'], decimals = 8)

    EPA_obs_df = EPA_obs_df.drop(columns = ['State Code','County Code','Site Num','Parameter Code','POC','Datum','Sample Duration','Date Local', 'Event Type',
           'Observation Count', 'Observation Percent','1st Max Value', '1st Max Hour', 'Address', 'County Name', 'City Name',
           'CBSA Name', 'Date of Last Change', 'Method Name'])

    #only use the SO2 1 hour 2010 pollutant standard arithmetic means and drop the 3-hour 1971 arithmetic means
    EPA_obs_df = EPA_obs_df.loc[~(EPA_obs_df['Pollutant Standard'] == 'SO2 3-hour 1971')]
    EPA_obs_df = EPA_obs_df.rename(columns = {'Parameter Name':'species'})

    #convert Ozone to ppb
    EPA_obs_df.loc[EPA_obs_df['species'] == 'Ozone','Arithmetic Mean'] *= 1e3 #ppb
    EPA_obs_df = EPA_obs_df.loc[~(EPA_obs_df['Arithmetic Mean'] <= 0)]
    
    #rename pollutants to shortened names
    for species in ['PM25', 'SO2', 'NO2', 'O3']:
        EPA_obs_df.loc[(EPA_obs_df['species'] == species_dict[species]), 'species'] = species
    
    #limit to just contiguous US (leaving out HI and AL)
    EPA_obs_df = EPA_obs_df.loc[(EPA_obs_df['Latitude'].between(24,50,inclusive = True)) & (EPA_obs_df['Longitude'].between(-130,-60,inclusive = True))]

    return(EPA_obs_df)
```

```python
def obs_model_plot(ds, df, species, 
                   vmin, vmax, cmap, figsize, month,convert_ppb,convert_ppb_df,
                   lat_lon, lat_spacing=47,lon_spacing=73 
                   ):
    fig, ax =  plt.subplots(figsize=figsize,subplot_kw={'projection':ccrs.LambertConformal()})
    ###### Create axes ######
    ax.coastlines()
    ax.add_feature(cfeat.STATES)
    PCM_m = ds.groupby('time.month').sum(dim = 'time').sel(month = month).isel(lev = 0)[f'{species}']
    if convert_ppb == True:
        PCM_m*=1e9
    
    ####### GEOS-CHEM output #######
    #PCM parameters and plot for model
    (PCM_m).plot(
        cmap = cmap, transform=ccrs.PlateCarree(), vmin = vmin, vmax = vmax, add_colorbar = False )
    
    ###### observations #######
    #create lat and lon for observations
    lat_o = df.loc[(df['species'] == species) & (df.date.dt.month == month)]['Latitude'].unique()
    lon_o = df.loc[(df['species'] == species) & (df.date.dt.month == month)]['Longitude'].unique()
    #define the concentrations for observations
    mean_conc=df.loc[(df['species'] == species) & (df.date.dt.month == month)].groupby(['Latitude','Longitude']).mean()['Arithmetic Mean']
    if convert_ppb_df == True:
        mean_conc*=1e9
    #PCM parameters and plot for observations
    PCM_o=ax.scatter(lon_o, lat_o, c=mean_conc, 
                     transform=ccrs.PlateCarree(),cmap=cmap,edgecolors='k',linewidth=.3,
                     s = 20,
                     vmin=vmin, vmax=vmax)
    plt.colorbar(PCM_o, ax=ax,extend='max', shrink=.3) 
    
    ###### adjustments and labels ########
    #adjust lat&lon being mapped
    ax.set_extent(lat_lon)
    plt.title(f'{month} {species}'); #title
```

```python
months = ['01','02','03','04','05','06','07','08','09','10','11','12']
```

```python
def interp_function(obs_df, ds):
    ### interpolate data in nearest method
    interp_EPA_df = pd.DataFrame(columns=['Arithmetic Mean', 'Longitude', 'Latitude','species','date'])

    for species in ['PM25', 'SO2', 'NO2', 'O3']:
        for month in pd.DatetimeIndex(obs_df.loc[obs_df.date.dt.month.isin(months)]['date'].values).month.unique().values:

            #data selected for date
            data = ds[f'{species}'].groupby('time.month').sum(dim = 'time').sel(month = month).isel(lev = 0)

            #new lat and lon in radians
            lats_new = obs_df.loc[(obs_df['species'] == species)]['Latitude'].unique()
            lons_new = obs_df.loc[(obs_df['species'] == species)]['Longitude'].unique()

            #interpolation function
            interp_data = []
            for idx in range(lats_new.size):
                interp_data.append(data.sel(lat=lats_new[idx], lon=lons_new[idx], method='nearest').values.item())
            tmp_df = pd.DataFrame({'Arithmetic Mean':interp_data, 'Longitude':lons_new, 'Latitude':lats_new, 'species': species, 'date': month})
            interp_EPA_df = interp_EPA_df.append(tmp_df, sort=False, ignore_index=True)

    for i in range(0, len(interp_EPA_df)):
        interp_EPA_df.loc[i,('date')] = datetime(2016,interp_EPA_df['date'][i],calendar.monthrange(2016,interp_EPA_df['date'][i])[1])  
    return(interp_EPA_df)
```

```python
interp_dev = interp_function(EPA_obs_df, devdata)
interp_ref = interp_function(EPA_obs_df, refdata)

    
```

```python
def mol_mol_ppb(interp_EPA_df):
    interp_EPA_df.loc[(interp_EPA_df['species'] == 'O3'),'Arithmetic Mean'] = interp_EPA_df.loc[(interp_EPA_df['species'] == 'O3')]['Arithmetic Mean'].apply(lambda x: x/1.0E9) 
    interp_EPA_df.loc[(interp_EPA_df['species'] == 'SO2'),'Arithmetic Mean'] = interp_EPA_df.loc[(interp_EPA_df['species'] == 'SO2')]['Arithmetic Mean'].apply(lambda x: x/1.0E9)
    interp_EPA_df.loc[(interp_EPA_df['species'] == 'NO2'),'Arithmetic Mean'] = interp_EPA_df.loc[(interp_EPA_df['species'] == 'NO2')]['Arithmetic Mean'].apply(lambda x: x/1.0E9)
    return(interp_EPA_df)
interp_dev = mol_mol_ppb(interp_dev)
interp_ref = mol_mol_ppb(interp_ref)
```

```python
interp_ref.to_csv('interp_ref_mod')
interp_dev.to_csv('interp_dev_mod')
```

```python
def linregress_data(obs_df, interp_df, month_string, species_list):
    result = [stats.linregress(
            x = obs_df.loc[
            (obs_df['species'] == species)
            ].groupby(['Latitude','Longitude']).mean()['Arithmetic Mean'],
            y = interp_df.loc[
            (interp_df['species'] == species)
            ].groupby(['Latitude','Longitude']).mean()['Arithmetic Mean']
            ) for species in species_list]
    lin_regress_df = pd.merge(pd.DataFrame([(species) for species in species_list], columns = ['species']),
             pd.DataFrame(result), 
             right_index = True, 
             left_index = True)
    return lin_regress_df

lin_regress_ref_df = linregress_data(EPA_obs_df, interp_ref, months, ['PM25', 'SO2', 'NO2', 'O3'])
lin_regress_dev_df = linregress_data(EPA_obs_df, interp_dev, months, ['PM25', 'SO2', 'NO2', 'O3'])
```

```python
def interp_scatterplot(interp_dfs, obs_df, lin_regress_df,species_list, month_string, rows,columns):
    fig, axes = plt.subplots(rows,columns,figsize = [6,12])
    for idx_m, interp_df in enumerate(interp_dfs):
        for idx_s, species in enumerate(species_list):
            ax = axes[idx_s, idx_m]
            x = obs_df.loc[(obs_df['species'] == species)].groupby(['Latitude','Longitude']).mean()['Arithmetic Mean']
            y = interp_df.loc[(interp_df['species'] == species)].groupby(['Latitude','Longitude']).mean()['Arithmetic Mean']
            ax.scatter(x, y, marker = '.')
            ax.plot(x, x + 0, 'xkcd:grey', label = '1:1 Line')
            #ax.plot(x,abline,'xkcd:almost black' , label = 'Linear Regression')
            ax.set_xlabel('Observational Annual Mean')
            ax.set_ylabel('GC Annual Mean')
            ax.set_xlim([-1,x.max()+1])
            ax.set_ylim([-1,x.max()+1])
            r_val = np.round(lin_regress_df.loc[(lin_regress_df['species'] == species), 'rvalue'].values[0], 2)
            std_err = np.round(lin_regress_df.loc[(lin_regress_df['species'] == species), 'stderr'].values[0], 2)
            ax.set_title(f'{species} \n R-value: {r_val} \n Standard error: {std_err}')
        custom_lines = [Line2D([0], [0], color='xkcd:grey', lw=4),
                    Line2D([0], [0], color='xkcd:almost black', lw=4)]
        plt.legend(custom_lines, ['1:1 Line', 'Linear Regression'])
        plt.tight_layout()
```
